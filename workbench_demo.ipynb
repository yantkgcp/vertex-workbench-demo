## PySpark

```python
from pyspark.sql import SparkSession

spark = SparkSession \
  .builder \
  .appName('spark-bigquery') \
  .getOrCreate()

# Use the Cloud Storage bucket for temporary BigQuery export data used
# by the connector.
bucket = "yantk-michigan-nb-staging"
spark.conf.set('temporaryGcsBucket', bucket)

# Load data from BigQuery.
location = spark.read.format('bigquery') \
  .option('table', 'michigan-351201.covid_siads591.location') \
  .load()

timeseries = spark.read.format('bigquery') \
  .option('table', 'michigan-351201.covid_siads591.timeseries') \
  .load()

sdf = timeseries.join(location, on='id', how='left')
```

```python
import pyspark.sql.functions as F

sdf = sdf.filter(sdf.population.isNotNull()).filter(sdf.deaths.isNotNull())

split = sdf.groupby('administrative_area_level_1')
apply = split.agg((F.max('deaths')/F.max('population')).alias("death_rate_pct"))
s_combine = apply.sort(F.desc("death_rate_pct"))
s_combine.show() ## 20 highest death rate countries
```

```python
import base64
import json
from datetime import datetime
from time import sleep
from google.cloud.notebooks_v1.types import schedule
from google.cloud.notebooks_v1.types import service
from google.cloud.notebooks_v1.services.notebook_service import NotebookServiceClient


def nb_schedule():
    """Triggered by a change to a Cloud Storage bucket.
    Args:
         event (dict): Event payload.
         context (google.cloud.functions.Context): Metadata for the event.
         sample: {"name": <pipeline name>, "notebook_path": <gcs nb path>, "outputNotebookFolder": <gcs path>}
    """
    client = NotebookServiceClient()
    TIMESTAMP = datetime.now().strftime("%Y%m%d%H%M%S")
    pipeline_name = 'demo_pipeline'
    notebook_path = 'gs://yantk-michigan-nb-staging/destination/pipeline/pipeline_at_2022_07_09_14_00_00.ipynb'
    outputNotebookFolder = 'gs://yantk-michigan-nb-staging/destination' + '/pipeline'
    name = 'projects/michigan-351201/locations/us-west1/schedules/' + pipeline_name + '_' + TIMESTAMP
    displayName = pipeline_name + '_' + TIMESTAMP

    #notebook_path = 'gs://cloud-ai-platform-215613aa-2b9d-497b-ae76-dee1c0608dd6/executor_files/checkride_automl_pipeline__1643094590136/checkride_automl_pipeline.ipynb'
    containerImageUri = 'gcr.io/deeplearning-platform-release/base-cu113:latest'

    ExecutionTemplate = {
         "scale_tier": 'CUSTOM',
         "master_type": 'n1-standard-4',
         "input_notebook_file": notebook_path,
         "container_image_uri": containerImageUri,
         "output_notebook_folder": outputNotebookFolder,
         "job_type": 'VERTEX_AI',
         "service_account": 'vertex-managed-notebook@michigan-351201.iam.gserviceaccount.com'
     }

    sch = {'cron_schedule': "00 18 * * 5", 'time_zone': 'Asia/Hong_Kong', 'execution_template': ExecutionTemplate}
    request = {'parent': 'projects/michigan-351201/locations/us-west1', 'schedule_id': displayName, 'schedule': sch}

    client.create_schedule(request)

    sleep(1.00)

    client.get_schedule(name=name)
```
